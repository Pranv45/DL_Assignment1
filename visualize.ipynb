```python
import matplotlib.pyplot as plt
from data.dataset import FashionMNISTLoader

loader = FashionMNISTLoader()
(X_train, y_train), _ = loader.get_data(normalize=False, flatten=False)
class_names = loader.get_class_names()
```

```python
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
axes = axes.flatten()

for i in range(10):
    idx = (y_train == i).argmax()
    axes[i].imshow(X_train[idx], cmap='gray')
    axes[i].set_title(class_names[i])
    axes[i].axis('off')

plt.suptitle("Fashion-MNIST: One Sample per Class", fontsize=16)
plt.tight_layout()
plt.show()
```
